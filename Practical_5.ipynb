**Aim: Write a panda program to split a given dataframe into groups with multiple aggregations.**


```python
import pandas as pd
#pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
df=pd.DataFrame({"school_code":["s001","s002","s003","s001","s002","s001"],
                 "class":["V","V","VI","VI","V","VI"],
                 "name":["Atharv","Chaitanya","Omkar","Dinkar","Bharat","Sagar"],
                 "DOB":["13/09/2003","04/02/2003","19/02/1920","10/07/2004","20/10/1990","5/08/1900"],
                 'age':[12,12,13,13,14,12],'height':[180,192,186,167,151,159],
                 'weight':[35,32,33,30,31,32],
                 'address':['street1','street2','street3','street4','street5','street6']},
                 index=['S1','S2','S3','S4','S5','S6',])
print("Original DataFrame:")
print("Group by with multiple aggregations:")
result=df.groupby(["school_code","class"]).agg({"height":["max","mean"],"weight":["sum","mean","count"]})
print(df)
```
