```python
#REMOVING PUNCTUTAION
import re
for k in data.keys():
  data[k]=re.sub(r'[-./?!,";:()\')]','',data[k])
print(data['IOT'][:800])
```

```python
#REMOVING NUMBER:
for k in data.keys():
  data[k]=re.sub('[-|0-9]','',data[k])
print(data['IOT'][:800])
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
# removing stopwords
for k in data.keys():
  print(data[k])
  data[k] = data[k].split()
  print(data[k])
  stopwords_list = stopwords.words('english')
  stopwords_list = stopwords_list +['mad','max','flim','fury']
  for k in data.keys():
    data[k] = [w for w in data[k] if not w in stopwords_list]
```

```python
from nltk.corpus.reader.tagged import word_tokenize
nltk.download('punkt')
example="this is an example showing off stop words"
stop_words = set(stopwords.words("english"))
filtered_sent=[]
words=word_tokenize(example)
```

```python
for w in words:
  print(w)
  if w not in stop_words:
    filtered_sent.append(w)
  print(filtered_sent)
```
