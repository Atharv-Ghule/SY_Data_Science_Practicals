```python
import seaborn as sns
sns.get_dataset_names()
```

```python
data=sns.load_dataset('titanic')
data.head()
```

```python
df_dropped=data.drop('survived',axis=1)
df_dropped.head()
```

```python
data.iloc[:5,0]
```

```python
data.loc[:5,'fare']
```

```python
df_dup=data.copy()
#duplicate the first row and append it to the data
row=df_dup.iloc[:1]
df_dup=df_dup.append(row,ignore_index=True)
df_dup
```

```python
df_dup[df_dup.duplicated()]
```

```python
data[data['pclass']==1]
```

```python
data[data['pclass'].isin([1,0])]
```

```python
data.groupby('sex').agg({'fare': 'count'})
```

```python
data.groupby('sex').agg({'fare': 'mean'})
```

```python
data.groupby(['pclass','sex']).agg({'fare':'count'})
```
